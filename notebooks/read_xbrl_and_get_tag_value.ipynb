# Read XBRL and Get Tag Value

XBRLファイルを読み込み、特定タグのデータを取得する


```python
import os
from pathlib import Path
import edinet

# サントリーホールディングス株式会社/H31.03.26 12:50
DOC_ID = "S100FGSC"

# Data Folder
DATA_ROOT = Path.cwd().joinpath("data")
```

## Download XBRL

```python
from edinet.xbrl_file import XBRLDir


# Download and load document
xbrl_path = edinet.api.document.get_xbrl(
    DOC_ID, save_dir=DATA_ROOT.joinpath("raw"),
    expand_level="dir")

xbrl_dir = XBRLDir(xbrl_path)
```

## Find tag and get value



```python
xbrl_dir.xbrl.find("jpcrp_cor:CompanyNameCoverPage").text
```

```python
xbrl_dir.xbrl.find("jpdei_cor:DocumentTypeDEI").text
```
